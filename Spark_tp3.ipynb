```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
rdd1 = sc.textFile('Shakespeare.txt')
```

```python
rdd1.collect()
```

```python
rdd2 = rdd1.flatMap(lambda rdd1: rdd1.split(" "))
```

```python
print(rdd2)
```

```python
rdd3 = rdd2.map(lambda mot: (1, mot))
```

```python
rdd3.collect()
```

```python
rdd4= rdd3.reduceByKey(lambda x,y : x+y)
```

```python
rdd4.saveAsTextFile('/content/r1')
```

```python
rdd4.coalesce(1).saveAsTextFile('/content/r3')
```
